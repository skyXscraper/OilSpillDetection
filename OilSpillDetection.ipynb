#K-means & TLS

```python
def enhance_image_tls_kmeans(image_path, K=3):
    img = cv2.imread(image_path)
    hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    stretched_hsv = np.zeros_like(hsv_img)
    for i in range(3):
        channel = hsv_img[:, :, i]
        low_p = np.percentile(channel, 2)
        high_p = np.percentile(channel, 98)
        channel_clipped = np.clip(channel, low_p, high_p)
        stretched = ((channel_clipped - low_p) * 255 / (high_p - low_p))
        stretched_hsv[:, :, i] = stretched.astype(np.uint8)

    Z = stretched_hsv.reshape((-1, 3)).astype(np.float32)

    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
    _, labels, centers = cv2.kmeans(Z, K, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

    centers = np.uint8(centers)
    segmented_data = centers[labels.flatten()].reshape(hsv_img.shape)

    return segmented_data ## applying TLS first, and then Kmeans
```

```python
from tqdm import tqdm
```

```python
def preprocess_dataset(input_folder, output_folder):
    os.makedirs(output_folder, exist_ok=True)
    for filename in tqdm(os.listdir(input_folder)):
        if filename.lower().endswith((".jpg", ".png", ".jpeg")):
            input_path = os.path.join(input_folder, filename)
            output_path = os.path.join(output_folder, filename)

            processed_img = enhance_image_tls_kmeans(input_path)
            cv2.imwrite(output_path, cv2.cvtColor(processed_img, cv2.COLOR_HSV2BGR))
## function to preprocess the entire data
```

```python
## preprocessing the train,val and test imgs

preprocess_dataset('/content/oil-spill-yolov8-complete-dataset-1/test/images', '/content/oil-spill-yolov8-complete-dataset-1/test/processed_images')
preprocess_dataset('/content/oil-spill-yolov8-complete-dataset-1/train/images', '/content/oil-spill-yolov8-complete-dataset-1/train/processed_images')
preprocess_dataset('/content/oil-spill-yolov8-complete-dataset-1/valid/images', '/content/oil-spill-yolov8-complete-dataset-1/valid/processed_images')
```
